```python
from selenium import webdriver as wd
from webdriver_manager.chrome import ChromeDriverManager

import random
import time
import pandas as pd
from bs4 import BeautifulSoup
```

```python
wd = wd.Chrome(ChromeDriverManager().install())
```

```python
wd.get("https://www.bestbuy.com/site/searchpage.jsp?id=pcat17071&qp=gpusv_facet%3DGraphics%20Processing%20Unit%20(GPU)~NVIDIA%20GeForce%20RTX%203060%5Egpusv_facet%3DGraphics%20Processing%20Unit%20(GPU)~NVIDIA%20GeForce%20RTX%203060%20Ti&st=gpu")
wd.implicitly_wait(15)
```

```python
def checkForStock(url):
    # soup = BeautifulSoup(wd.page_source)
    wd.get(url)
    soup = BeautifulSoup(wd.page_source,features="html.parser")
    items = soup.find("div", {"class": "component-sku-list"})

    items_processed = []

    for row in items.findAll("div", {"class": "list-item lv"}):
        row_processed = []
        itemTitle = row.find("h4", {"class": "sku-header"})
        itemURL = row.find("a")
        itemPromoText = row.find("button", {"class": "c-button"})
        itemPrice = row.find("div",{"class": "priceView-customer-price"})
        status = "Available"

        if itemPromoText and itemPromoText.text == "Sold Out":
            status = "Sold Out"

        if itemTitle:
            row_processed.append(itemTitle.text)
            row_processed.append(itemPrice.find("span").text)
            row_processed.append(itemURL.get("href"))
            row_processed.append(status)

        if row_processed:
            items_processed.append(row_processed)
    # cells[3].find("img"valid)["src"]
    # columns = ["ImageUrl","Origin"]

    df = pd.DataFrame.from_records(items_processed, columns=["Item Title", "Item Price", "URL", "Status"])

    df["Item Price"] = df["Item Price"].apply(lambda x: x.replace("$",""))
    df["Item Price"] = df["Item Price"].apply(lambda x: x.replace(".",""))
    df["Item Price"] = pd.to_numeric(df["Item Price"])
    
    return df
```
