```python
from selenium import webdriver as wd
from webdriver_manager.chrome import ChromeDriverManager

import random
import time
```

```python
wd = wd.Chrome(ChromeDriverManager().install())
wd.implicitly_wait(15)

wd.get("https://www.bestbuy.com/site/pny-xlr8-gaming-single-fan-nvidia-geforce-gtx-1660-super-overclocked-edition-6gb-gddr6-pci-express-3-0-graphics-card-black/6407309.p?skuId=6407309")
```

```python
random_wait_time = random.randrange(5.0,10.0)
print(random_wait_time)
time.sleep(random_wait_time)
try:
    add_to_cart_button = wd.find_element_by_xpath('/html/body/div[3]/main/div[2]/div[3]/div[2]/div/div/div[7]/div[1]/div/div/div/button')
    add_to_cart_button.click()
except NoSuchElementException:
    print("Out of Stock")
```
