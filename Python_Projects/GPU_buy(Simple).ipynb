```python
from selenium import webdriver as wd
from webdriver_manager.chrome import ChromeDriverManager

import random
import time
```

```python
wd = wd.Chrome(ChromeDriverManager().install())
```

```python
wd.get("https://www.bestbuy.com/site/nvidia-geforce-rtx-3060-ti-8gb-gddr6-pci-express-4-0-graphics-card-steel-and-black/6439402.p?skuId=6439402")
wd.implicitly_wait(15)
```

```python
from selenium.common.exceptions import NoSuchElementException

random_wait_time = random.randrange(5.0,10.0)
print(random_wait_time)
time.sleep(random_wait_time)

add_to_cart_button = wd.find_element_by_xpath('/html/body/div[3]/main/div[2]/div[3]/div[2]/div/div/div[14]/div[2]/div/div/div/button')
add_to_cart_button.click()

while (True):
    try:
        wd.get("https://www.bestbuy.com/site/nvidia-geforce-rtx-3060-ti-8gb-gddr6-pci-express-4-0-graphics-card-steel-and-black/6439402.p?skuId=6439402")
        add_to_cart_button = wd.find_element_by_xpath('/html/body/div[3]/main/div[2]/div[3]/div[2]/div/div/div[14]/div[2]/div/div/div/button')
        add_to_cart_button.click()
        break
    except NoSuchElementException:
        print("Out of Stock")
        print("Waiting to Try Again")
        time.sleep(120)
```
